# N =5 Symmetric

```python
file_name='./fitness_vals/cycle5datafile_2024-09-03_11-11-23.csv'

with open(file_name) as csvfile:
    generations =[]#define new generation list
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        generations.append(row)
        #print(generations)
        
num_generations = len(generations)#total number of generation
num_elements = len(generations[0])#total number of individuals in the population

elements, element_list =Plotter.setup_data_arrays_all(generations, num_generations, num_elements)



#Now we plot
plt.plot(element_list,elements)
plt.title("N=5 cycle graph, S =1 check later, E =0 \n"
         "1 Move, No waiting, Symmetric angles")
plt.xlabel("Solution number")
plt.ylabel("Probability of winning the game")

counter =0 
while counter<num_generations:
    plt.axvline(counter*(num_elements-2), color = "black", linestyle ="--", alpha = 0.7)
    counter = counter +1
```

```python
file_name='./showing_computational_convergence/cycle3-1millionruns.csv'

with open(file_name) as csvfile:
    generations =[]#define new generation list
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        generations.append(row)
        #print(generations)
        
num_generations = len(generations)#total number of generation
num_elements = len(generations[0])#total number of individuals in the population

elements, element_list =Plotter.setup_data_arrays_maxval(generations, num_generations, num_elements)


fig, ax2 = plt.subplots()


#Now we plot
plt.plot(element_list,elements)
plt.title("N=3 cycle graph, S =1 check later, E =0 \n"
         "1 Move, No waiting, Symmetric angles")
plt.xlabel("Generation Number")
plt.ylabel("Probability of winning the game")
plt.axhline(y =0.58333, label = "Analytical Quantum Bound", linestyle = '--', color = 'black')
plt.axhline(y =0.555555, label = "Analytical Classical Bound",linestyle = '--', color = 'grey')
plt.legend(loc='center right')


##Now do inset


left, bottom, width, height = [0.4, 0.15,0.5, 0.23]#
ax2 = fig.add_axes([left, bottom, width, height]) 
ax2.axhline(y=0.583, color='black', linestyle='--', linewidth=2,label='Theoretical Quantum Value')
plt.xlim(75,100)
plt.ylim(0.57,0.59)
ax2.plot(element_list,elements, 'o-',color='red',alpha=0.75)

plt.savefig("3CycleGAS1Symmetric.eps", format = "eps")
plt.savefig("3CycleGAS1Symmetric.png", format = "png")
```

Now lets plot both the classical strategy and the quantum strategy on the same graph

```python
file_name='./showing_computational_convergence/cycle3-1millionruns.csv'
file_name2 = './fitness_vals/cycle3_classicaldatafile_2024-09-23_13-15-21.csv'
with open(file_name) as csvfile:
    generations =[]#define new generation list
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        generations.append(row)
        #print(generations)
        
with open(file_name2) as csvfile:
    generations2 =[]#define new generation list
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        generations2.append(row)
        #print(generations)
        
num_generations = len(generations)#total number of generation
num_elements = len(generations[0])#total number of individuals in the population

elements, element_list =Plotter.setup_data_arrays_maxval(generations, num_generations, num_elements)

elements2, element_list2 =Plotter.setup_data_arrays_maxval(generations2, num_generations, num_elements)

fig, ax2 = plt.subplots()


#Now we plot
plt.plot(element_list,elements, label = 'Quantum Strategy')
plt.plot(element_list2,elements2, label = 'Classical Strategy')
plt.title("N=3 cycle graph, S =1 check later, E =0 \n"
         "1 Move, No waiting, Symmetric angles")
plt.xlabel("Generation Number")
plt.ylabel("Probability of winning the game")
plt.axhline(y =0.58333, label = "Analytical Quantum Bound", linestyle = '--', color = 'black')
plt.axhline(y =0.555555, label = "Analytical Classical Bound",linestyle = '--', color = 'grey')
plt.legend(loc='center right')


##Now do inset


left, bottom, width, height = [0.4, 0.15,0.5, 0.23]#
ax2 = fig.add_axes([left, bottom, width, height]) 
ax2.axhline(y=0.583, color='black', linestyle='--', linewidth=2,label='Theoretical Quantum Value')
plt.axhline(y =0.555555, label = "Analytical Classical Bound",linestyle = '--', color = 'grey')
plt.xlim(75,100)
plt.ylim(0.54,0.59)
ax2.plot(element_list,elements, 'o-',color='b',alpha=0.75)
ax2.plot(element_list2,elements2, 'o-',color='orange',alpha=0.75)

plt.savefig("3CycleGAComparisonSymmetric.eps", format = "eps")
plt.savefig("3CycleGAComparionSymmetric.png", format = "png")
```
